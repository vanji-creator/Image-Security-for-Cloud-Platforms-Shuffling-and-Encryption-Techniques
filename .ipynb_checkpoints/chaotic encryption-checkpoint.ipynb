```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Load an image
image_path = r"C:\Users\Asus\python image encryption\images\pirate.tif"
original_image = cv2.imread(image_path)

# Flatten the original image to a 1D array for each channel
original_image_1d = original_image.reshape((-1, 3))

# Calculate the number of pixels
num_pixels = len(original_image_1d)

# Define the Tinkerbell map equations
def tinkerbell_map(x, y, a, b, c, d):
    x_new = x**2 - y**2 + a * x + b * y
    y_new = 2 * x * y + c * x + d * y
    return x_new, y_new

# Generate a chaotic sequence for each channel using the Tinkerbell map
a = 0.9
b = -0.6013
c = 2.0
d = 0.5
chaotic_sequence_x = np.zeros(num_pixels)
chaotic_sequence_y = np.zeros(num_pixels)
x, y = 0.1, 0.1  # Initial conditions
for i in range(num_pixels):
    x, y = tinkerbell_map(x, y, a, b, c, d)
    chaotic_sequence_x[i] = x
    chaotic_sequence_y[i] = y

# Multiply each pixel value by its assigned chaotic sequence value for each channel
encrypted_image_1d = original_image_1d.astype(float) * chaotic_sequence_x[:num_pixels, np.newaxis]
encrypted_image_1d *= chaotic_sequence_y[:num_pixels, np.newaxis]

# Reshape the encrypted 1D array back to the original image shape
encrypted_image = encrypted_image_1d.reshape(original_image.shape)

# Display the original image
plt.figure(figsize=(6, 6))
plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.show()

# Display the encrypted image
plt.figure(figsize=(6, 6))
plt.imshow(cv2.cvtColor(encrypted_image.astype(np.uint8), cv2.COLOR_BGR2RGB))
plt.title('Encrypted Image')
plt.show()

# Save the encrypted image
# cv2.imwrite(r'C:\Users\Asus\python image encryption\tinkerbell_encrypted_image.jpg', encrypted_image)
# print("Encrypted image saved as 'tinkerbell_encrypted_image.jpg'")

# Decryption: Divide each pixel by its assigned chaotic sequence value for each channel
decrypted_image_1d = encrypted_image_1d / chaotic_sequence_x[:num_pixels, np.newaxis]
decrypted_image_1d /= chaotic_sequence_y[:num_pixels, np.newaxis]

# Reshape the decrypted 1D array back to the original image shape
decrypted_image = decrypted_image_1d.reshape(original_image.shape)

# Display the decrypted image
plt.figure(figsize=(6, 6))
plt.imshow(cv2.cvtColor(decrypted_image.astype(np.uint8), cv2.COLOR_BGR2RGB))
plt.title('Decrypted Image')
plt.show()

# Save the decrypted image
# cv2.imwrite(r'C:\Users\Asus\python image encryption\tinkerbell_decrypted_image.jpg', decrypted_image)
# print("Decrypted image saved as 'tinkerbell_decrypted_image.jpg'")
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Load an image
image_path = r"C:\Users\Asus\python image encryption\images\pirate.tif"
original_image = cv2.imread(image_path)

# Flatten the original image to a 1D array for each channel
original_image_1d = original_image.reshape((-1, 3))

# Calculate the number of pixels
num_pixels = len(original_image_1d)

# Define the Tinkerbell map equations
def tinkerbell_map(x, y, a, b, c, d):
    x_new = x**2 - y**2 + a * x + b * y
    y_new = 2 * x * y + c * x + d * y
    return x_new, y_new

# Generate a chaotic sequence for each channel using the Tinkerbell map
a = 0.9
b = -0.6013
c = 2.0
d = 0.5
chaotic_sequence_x = np.zeros(num_pixels)
chaotic_sequence_y = np.zeros(num_pixels)
x, y = 0.1, 0.1  # Initial conditions
for i in range(num_pixels):
    x, y = tinkerbell_map(x, y, a, b, c, d)
    chaotic_sequence_x[i] = x
    chaotic_sequence_y[i] = y

# Sort the chaotic sequences and get the permutation indices
sorted_indices_x = np.argsort(chaotic_sequence_x)
sorted_indices_y = np.argsort(chaotic_sequence_y)

# Rearrange the pixels of the image based on the sorted order of the sequences
encrypted_image_1d = original_image_1d[sorted_indices_x]
encrypted_image_1d = encrypted_image_1d[sorted_indices_y]

# Reshape the encrypted 1D array back to the original image shape
encrypted_image = encrypted_image_1d.reshape(original_image.shape)

# Display the original image
plt.figure(figsize=(6, 6))
plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.show()

# Display the encrypted image
plt.figure(figsize=(6, 6))
plt.imshow(cv2.cvtColor(encrypted_image.astype(np.uint8), cv2.COLOR_BGR2RGB))
plt.title('Encrypted Image')
plt.show()

# Save the encrypted image
cv2.imwrite(r'C:\Users\Asus\phase 2 project\tinkerbell_encrypted_image.jpg', encrypted_image)
print("Encrypted image saved as 'tinkerbell_encrypted_image.jpg'")
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Load the encrypted image
encrypted_image_path = r'C:\Users\Asus\phase 2 project\tinkerbell_encrypted_image.jpg'
encrypted_image = cv2.imread(encrypted_image_path)

# Flatten the encrypted image to a 1D array for each channel
encrypted_image_1d = encrypted_image.reshape((-1, 3))

# Calculate the number of pixels
num_pixels = len(encrypted_image_1d)

# Define the Tinkerbell map equations
def tinkerbell_map(x, y, a, b, c, d):
    x_new = x**2 - y**2 + a * x + b * y
    y_new = 2 * x * y + c * x + d * y
    return x_new, y_new

# Generate the same chaotic sequences for each channel using the Tinkerbell map
a = 0.9
b = -0.6013
c = 2.0
d = 0.5
chaotic_sequence_x = np.zeros(num_pixels)
chaotic_sequence_y = np.zeros(num_pixels)
x, y = 0.1, 0.1  # Initial conditions

# Store the original order of the pixels
original_order = np.arange(num_pixels)

# Generate the chaotic sequences and store them
for i in range(num_pixels):
    x, y = tinkerbell_map(x, y, a, b, c, d)
    chaotic_sequence_x[i] = x
    chaotic_sequence_y[i] = y

# Sort the chaotic sequences to get the permutation indices
sorted_indices_x = np.argsort(chaotic_sequence_x)
sorted_indices_y = np.argsort(chaotic_sequence_y)

# Rearrange the pixels of the encrypted image based on the original order
decrypted_image_1d = encrypted_image_1d[original_order]

# Rearrange the pixels of the encrypted image based on the sorted order of the sequences
decrypted_image_1d = decrypted_image_1d[sorted_indices_x]
decrypted_image_1d = decrypted_image_1d[sorted_indices_y]

# Reshape the decrypted 1D array back to the original image shape
decrypted_image = decrypted_image_1d.reshape(encrypted_image.shape)

# Display the decrypted image
plt.figure(figsize=(6, 6))
plt.imshow(cv2.cvtColor(decrypted_image.astype(np.uint8), cv2.COLOR_BGR2RGB))
plt.title('Decrypted Image')
plt.show()

# Save the decrypted image
cv2.imwrite(r'C:\Users\Asus\phase 2 project\tinkerbell_decrypted_image.jpg', decrypted_image)
print("Decrypted image saved as 'tinkerbell_decrypted_image.jpg'")
```
